# ***Astropy Python Tutorial: How To Create a  Gaussain Model of 2D Disk Galaxy?***

### Step-1: import library

```python
from astropy.modeling.models import Gaussian2D
import numpy as np
import matplotlib.pyplot as plt
```

### Step-2: Create Galaxy pixel size

```python
x,=np.mgrid[0:300,0:200]
```

### create a  model galaxy

```python
model=Gaussian2D(1,100,100,50,50,np.pi/4)
model
```

```python
galaxy_model=model(x,y)
plt.imshow(galaxy_model,cmap='jet')
```
